```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tqdm
import tensorflow_datasets as tfds
from tensorflow import keras
```

```python
data_dir = 'data'
ds = tfds.load('voc/2012')
```

```python
classes = ['Aeroplanes', 'Bicycles', 'Birds', 'Boats', 'Bottles', 'Buses', 'Cars', 'Cats', 'Chairs','Cows', 
           'Dining tables', 'Dogs', 'Horses', 'Motorbikes', 'People', 'Potted plants', 'Sheep', 'Sofas',
            'Trains', 'TV/Monitors']
ds
```

```python
data_train = ds['train']
data_vals = ds['validation']
data_test = ds['test']
```

```python
assert isinstance(data_train, tf.data.Dataset)
data_test
```

```python
fig, ax = plt.subplots(1, 5, figsize=(15, 5))
for i, d in enumerate(data_train.take(5)):
    ax[i].imshow(d['image'].numpy())
    name_class = [classes[ind] for ind in d['labels']]
    ax[i].set_title(', '.join(name_class))
```

```python
data_train.element_spec
```

```python
def func_one_hot(x):
    y = x.numpy()
    h = np.zeros(20)
    h[y] = 1
    return h

@tf.function
def tf_func_one_hot(r):
    h = tf.py_function(func_one_hot, inp=[r], Tout=[tf.int32])
    return h[0]

for d in data_train.take(2):
# #     print(d['labels'])
      h = tf_func_one_hot(d['labels'])
      print(h)
#     h = func_one_hot(d['labels'])
#     print(d['labels'].numpy())
```

```python
size = (300, 300)
data_train_clf = data_train.map(lambda x: (tf.image.resize(x['image'], size), tf_func_one_hot(x['labels'])))


data_vals_clf = data_vals.map(lambda x: (tf.image.resize(x['image'], size), tf_func_one_hot(x['labels'])))

data_test_clf = data_test.map(lambda x: (tf.image.resize(x['image'], size), tf_func_one_hot(x['labels'])))
```

```python
# data_train_clf = data_train.map(lambda x: (x['image'], x['labels']))
data_train_clf.element_spec
# size = (15, 150)
# data_train_clf_resize = data_train_clf.map(lambda img: tf.image.resize(img[0], size))
# next(iter(data_train_clf.take(1)))[1].shape
```

```python
# some data augmentation
data_train_clf = data_train_clf.repeat(200)

# data_train_clf = data_train_clf.map(lambda image, label: (tf.image.random_flip_left_right(image), label))

# data_train_clf = data_train_clf.map(lambda image, label: (tf.image.random_flip_up_down(image), label))

# data_train_clf = data_train_clf.map(lambda image, label: (tf.image.random_hue(image, 0.2), label))

data_train_clf = data_train_clf.shuffle(buffer_size=1000)
```

```python
fig, ax = plt.subplots(1, 10, figsize=(20, 5))
i = 0
for img, label in data_train_clf.take(10):
    ax[i].imshow(img.numpy().astype(np.int32))
    i+=1
```

```python
data_train_clf = data_train_clf.batch(32)

data_vals_clf = data_vals_clf.batch(32)
data_test_clf = data_test_clf.batch(32)
```

```python
i = 0
for d in data_train_clf.take(1):
    i +=1
d[0].shape
```

```python
reg = tf.keras.regularizers.L1(0.0001)

model = keras.Sequential()
model.add(keras.layers.Rescaling(1./255, input_shape=(300, 300, 3)))
model.add(keras.layers.RandomFlip("horizontal_and_vertical"))
model.add(keras.layers.RandomRotation(.2))
# # model.add(keras.layers.RandomHeight(.2))
# # model.add(keras.layers.RandomWidth(.2))
model.add(keras.layers.RandomTranslation(height_factor=.2, width_factor=.2))
model.add(keras.layers.RandomZoom(.5, .2))


model.add(keras.layers.Conv2D(filters=16, kernel_size=(3, 3), activation='relu',kernel_regularizer=reg))    
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))

model.add(keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu',kernel_regularizer=reg))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))

model.add(keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu',kernel_regularizer=reg))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))

model.add(keras.layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu',kernel_regularizer=reg))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))
# model.add(keras.layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))

model.add(keras.layers.Conv2D(filters=128, kernel_size=(3, 3), activation='relu'))
model.add(keras.layers.MaxPooling2D(pool_size=(2, 2)))

model.add(keras.layers.Flatten())

model.add(keras.layers.Dropout(rate=.5))
model.add(keras.layers.Dense(256, activation='relu'))
model.add(keras.layers.Dense(20, activation='sigmoid'))
```

```python
model.summary()
```

```python
cce = tf.keras.losses.CategoricalCrossentropy()

def loss_func(label_true, lable_pred):
    return cce(label_true, lable_pred)
       
```

```python
model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.001),
              loss=loss_func)
```

```python
history = model.fit(data_train_clf,steps_per_epoch=300, validation_data=data_vals_clf, batch_size=32, epochs=100)
```
